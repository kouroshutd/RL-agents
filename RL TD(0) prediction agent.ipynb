```python
import numpy as np

n_states = 5
V = np.zeros(n_states)
alpha = 0.1
gamma = 1.0

for episode in range(1000):
    state = 2
    done = False

    while not done:
        next_state = state + np.random.choice([-1, 1])
        next_state = min(max(next_state, 0), n_states - 1)

        reward = 1 if next_state == n_states - 1 else 0
        done = next_state == 0 or next_state == n_states - 1

        V[state] += alpha * (reward + gamma * V[next_state] - V[state])
        state = next_state

print("Learned values:", V)
```
