```python
import os
import asyncio
from dotenv import load_dotenv
from agents import Agent, Runner

load_dotenv()

def get_api_key():
    key = os.getenv("OPENAI_API_KEY")
    if not key or key == "your_api_key_here":
        return None
    return key

agent = Agent(
    name="Assistant",
    instructions="You are a helpful assistant."
)

print("Setup loaded.")
```

```python
async def run_agent():
    api_key = get_api_key()

    if not api_key:
        print("OPENAI_API_KEY is missing or still a placeholder.")
        print("Your notebook environment is set up correctly.")
        print("Add a real key to .env later, then run again.")
        return

    try:
        result = await Runner.run(agent, "Say hello in one short sentence.")
        print(result.final_output)
    except Exception as e:
        print("The agent started, but the request failed.")
        print(f"Error: {e}")
```

```python
await run_agent()
```
