```python
import numpy as np

n_actions = 10
true_values = np.random.randn(n_actions)

q = np.zeros(n_actions)
counts = np.zeros(n_actions)
epsilon = 0.1
rewards = []

for step in range(1000):
    if np.random.rand() < epsilon:
        action = np.random.randint(n_actions)
    else:
        action = np.argmax(q)

    reward = np.random.randn() + true_values[action]
    counts[action] += 1
    q[action] += (reward - q[action]) / counts[action]
    rewards.append(reward)

print("Estimated values:", q)
print("Best learned action:", np.argmax(q))
```
