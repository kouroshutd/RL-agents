```python
import numpy as np

grid_size = 4
gamma = 0.9
V = np.zeros((grid_size, grid_size))
goal = (3, 3)

actions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def next_state(state, action):
    i, j = state
    ni = min(max(i + action[0], 0), grid_size - 1)
    nj = min(max(j + action[1], 0), grid_size - 1)
    return (ni, nj)

for _ in range(100):
    new_V = np.copy(V)
    for i in range(grid_size):
        for j in range(grid_size):
            if (i, j) == goal:
                continue
            values = []
            for a in actions:
                ns = next_state((i, j), a)
                reward = 0 if ns == goal else -1
                values.append(reward + gamma * V[ns])
            new_V[i, j] = max(values)
    V = new_V

print(V)
```
